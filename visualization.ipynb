```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the data
df = pd.read_csv('heart.csv')

# Set aesthetic style
plt.style.use('seaborn-v0_8-muted')

# Create a figure with a grid of subplots (3 rows, 2 columns)
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Heart Disease Data Analysis Dashboard', fontsize=24, fontweight='bold', y=0.95)

# 1. Distribution of Age
axes[0, 0].hist(df['age'], bins=15, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribution of Age', fontsize=14)
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

# 2. Heart Disease Prevalence (1 = No, 2 = Yes)
disease_counts = df['heart_disease'].value_counts().sort_index()
axes[0, 1].bar(['No Disease (1)', 'Has Disease (2)'], disease_counts, color=['lightgreen', 'salmon'])
axes[0, 1].set_title('Heart Disease Prevalence', fontsize=14)
axes[0, 1].set_ylabel('Count')

# 3. Serum Cholesterol vs Age (Colored by Heart Disease)
scatter = axes[1, 0].scatter(df['age'], df['serum_cholestoral'], c=df['heart_disease'], cmap='coolwarm', alpha=0.7)
axes[1, 0].set_title('Cholesterol vs Age', fontsize=14)
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Serum Cholesterol')
legend1 = axes[1, 0].legend(*scatter.legend_elements(), title="Heart Disease")
axes[1, 0].add_artist(legend1)

# 4. Chest Pain Type Distribution
cp_counts = df['chest_pain_type'].value_counts().sort_index()
axes[1, 1].bar(cp_counts.index.astype(str), cp_counts.values, color='plum')
axes[1, 1].set_title('Distribution of Chest Pain Types', fontsize=14)
axes[1, 1].set_xlabel('Chest Pain Type')
axes[1, 1].set_ylabel('Count')

# 5. Sex Distribution (Pie Chart)
sex_counts = df['sex'].value_counts()
axes[2, 0].pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'pink'])
axes[2, 0].set_title('Gender Distribution', fontsize=14)

# 6. Max Heart Rate Distribution
axes[2, 1].hist(df['max_heart_rate'], bins=15, color='orange', edgecolor='black')
axes[2, 1].set_title('Distribution of Max Heart Rate', fontsize=14)
axes[2, 1].set_xlabel('Max Heart Rate')
axes[2, 1].set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.93])

# Save the dashboard
plt.savefig('heart_disease_dashboard.png')

# Create a separate, more focused correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='RdBu', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.savefig('correlation_heatmap.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the data
df = pd.read_csv('diabetes_data_upload.csv')

# Set style
plt.style.use('seaborn-v0_8-whitegrid')

# Create dashboard figure
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Diabetes Risk Factor Analysis Dashboard', fontsize=24, fontweight='bold', y=0.98)

# 1. Age Distribution
sns.histplot(df['Age'], bins=15, kde=True, ax=axes[0, 0], color='teal')
axes[0, 0].set_title('Age Distribution of Patients', fontsize=14)

# 2. Class Distribution (Positive vs Negative)
class_counts = df['class'].value_counts()
axes[0, 1].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
axes[0, 1].set_title('Prevalence of Diabetes (Class)', fontsize=14)

# 3. Gender vs Class
sns.countplot(data=df, x='Gender', hue='class', ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Diabetes Diagnosis by Gender', fontsize=14)

# 4. Polyuria vs Class
sns.countplot(data=df, x='Polyuria', hue='class', ax=axes[1, 1], palette='magma')
axes[1, 1].set_title('Polyuria (Frequent Urination) vs Diabetes', fontsize=14)

# 5. Symptom Prevalence Summary (Percentage of 'Yes')
symptoms = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia', 
            'Genital thrush', 'visual blurring', 'Itching', 'Irritability', 
            'delayed healing', 'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity']

yes_percentages = [(df[col] == 'Yes').mean() * 100 for col in symptoms]
symptom_df = pd.DataFrame({'Symptom': symptoms, 'Percentage': yes_percentages}).sort_values('Percentage', ascending=False)

sns.barplot(data=symptom_df, x='Percentage', y='Symptom', ax=axes[2, 1], palette='Blues_r')
axes[2, 1].set_title('Prevalence of Symptoms (% Yes)', fontsize=14)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('diabetes_dashboard.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load both datasets
df_heart = pd.read_csv('heart.csv')
df_diabetes = pd.read_csv('diabetes_data_upload.csv')

# Set aesthetic style
plt.style.use('seaborn-v0_8-muted')

# Create a figure with a grid (4 rows, 2 columns)
fig, axes = plt.subplots(4, 2, figsize=(18, 22))
fig.suptitle('Combined Health Analysis Dashboard: Heart Disease & Diabetes', fontsize=26, fontweight='bold', y=0.98)

# --- HEART DISEASE SECTION (LEFT COLUMN) ---
sns.histplot(df_heart['age'], bins=15, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Heart: Age Distribution', fontsize=14)

heart_counts = df_heart['heart_disease'].value_counts().sort_index()
axes[1, 0].bar(['No Disease', 'Has Disease'], heart_counts, color=['lightgreen', 'salmon'])
axes[1, 0].set_title('Heart: Disease Prevalence', fontsize=14)

sns.histplot(df_heart['max_heart_rate'], bins=15, ax=axes[2, 0], color='orange', kde=True)
axes[2, 0].set_title('Heart: Max Heart Rate Achieved', fontsize=14)

cp_counts = df_heart['chest_pain_type'].value_counts().sort_index()
axes[3, 0].bar(cp_counts.index.astype(str), cp_counts.values, color='plum')
axes[3, 0].set_title('Heart: Chest Pain Type Counts', fontsize=14)

# --- DIABETES SECTION (RIGHT COLUMN) ---
sns.histplot(df_diabetes['Age'], bins=15, kde=True, ax=axes[0, 1], color='teal')
axes[0, 1].set_title('Diabetes: Age Distribution', fontsize=14)

dia_counts = df_diabetes['class'].value_counts()
axes[1, 1].pie(dia_counts, labels=dia_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
axes[1, 1].set_title('Diabetes: Prevalence', fontsize=14)

sns.countplot(data=df_diabetes, x='Gender', hue='class', ax=axes[2, 1], palette='viridis')
axes[2, 1].set_title('Diabetes: Diagnosis by Gender', fontsize=14)

symptoms = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia']
yes_percentages = [(df_diabetes[col] == 'Yes').mean() * 100 for col in symptoms]
symptom_df = pd.DataFrame({'Symptom': symptoms, 'Percentage': yes_percentages})
sns.barplot(data=symptom_df, x='Percentage', y='Symptom', ax=axes[3, 1], hue='Symptom', palette='Blues_r')
if axes[3, 1].get_legend() is not None: axes[3, 1].get_legend().remove()
axes[3, 1].set_title('Diabetes: Top Symptoms (% Yes)', fontsize=14)

plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.savefig('combined_health_dashboard.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load datasets
df_heart = pd.read_csv('heart.csv')
df_diabetes = pd.read_csv('diabetes_data_upload.csv')

# --- DATA PREPROCESSING ---
# Map Heart Disease labels
heart_label_map = {1: 'Healthy (No Disease)', 2: 'Heart Disease Positive'}
df_heart['heart_disease_label'] = df_heart['heart_disease'].map(heart_label_map)

# Map Chest Pain Types
cp_map = {1.0: 'Typical Angina', 2.0: 'Atypical Angina', 3.0: 'Non-Anginal', 4.0: 'Asymptomatic'}
df_heart['chest_pain_label'] = df_heart['chest_pain_type'].map(cp_map)

# --- PLOTTING ---
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({'font.size': 11, 'font.family': 'sans-serif'})

fig, axes = plt.subplots(4, 2, figsize=(20, 24))
fig.suptitle('Professional Medical Analytics Dashboard: Heart Disease & Diabetes Study', 
             fontsize=28, fontweight='bold', y=0.98, color='#2c3e50')

# Helper for adding bar annotations
def add_val_labels(ax, orient='v'):
    if orient == 'v':
        for p in ax.patches:
            height = p.get_height()
            if height > 0:
                ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                            ha='center', va='center', xytext=(0, 8), textcoords='offset points', fontweight='bold')
    else:
        for p in ax.patches:
            width = p.get_width()
            if width > 0:
                ax.annotate(f'{width:.1f}%', (width, p.get_y() + p.get_height() / 2.),
                            ha='left', va='center', xytext=(5, 0), textcoords='offset points', fontweight='bold')

# --- Column 1: Heart Metrics ---
sns.histplot(df_heart['age'], bins=12, kde=True, ax=axes[0, 0], color='#34495e', edgecolor='white')
axes[0, 0].set_title('A. Patient Age Distribution (Heart Study)', fontsize=15, fontweight='bold')

sns.countplot(data=df_heart, x='heart_disease_label', ax=axes[1, 0], palette=['#27ae60', '#c0392b'], order=['Healthy (No Disease)', 'Heart Disease Positive'])
axes[1, 0].set_title('B. Clinical Diagnosis Frequency', fontsize=15, fontweight='bold')
add_val_labels(axes[1, 0])

sns.kdeplot(data=df_heart, x='max_heart_rate', hue='heart_disease_label', fill=True, ax=axes[2, 0], palette=['#27ae60', '#c0392b'], alpha=0.4)
axes[2, 0].set_title('C. Heart Rate Capacity (BPM) by Condition', fontsize=15, fontweight='bold')

sns.countplot(data=df_heart, y='chest_pain_label', ax=axes[3, 0], palette='viridis')
axes[3, 0].set_title('D. Characterization of Chest Pain Types', fontsize=15, fontweight='bold')

# --- Column 2: Diabetes Analysis ---
sns.histplot(df_diabetes['Age'], bins=12, kde=True, ax=axes[0, 1], color='#16a085', edgecolor='white')
axes[0, 1].set_title('E. Patient Age Distribution (Diabetes Study)', fontsize=15, fontweight='bold')

dia_counts = df_diabetes['class'].value_counts()
axes[1, 1].pie(dia_counts, labels=dia_counts.index, autopct='%1.1f%%', startangle=90, colors=['#c0392b', '#2980b9'], shadow=True)
axes[1, 1].set_title('F. Positive vs. Negative Case Split', fontsize=15, fontweight='bold')

sns.countplot(data=df_diabetes, x='Gender', hue='class', ax=axes[2, 1], palette='Set2')
axes[2, 1].set_title('G. Demographic Risk (Gender Split)', fontsize=15, fontweight='bold')
add_val_labels(axes[2, 1])

symptoms_list = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia']
yes_perc = [(df_diabetes[col] == 'Yes').mean() * 100 for col in symptoms_list]
symptom_data = pd.DataFrame({'Symptom': [s.capitalize() for s in symptoms_list], 'Percentage': yes_perc}).sort_values('Percentage', ascending=False)
sns.barplot(data=symptom_data, x='Percentage', y='Symptom', ax=axes[3, 1], hue='Symptom', palette='mako')
if axes[3, 1].get_legend() is not None: axes[3, 1].get_legend().remove()
axes[3, 1].set_title('H. Key Indicator Prevalence (%)', fontsize=15, fontweight='bold')
add_val_labels(axes[3, 1], orient='h')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('professional_medical_dashboard.png', dpi=300)
```
